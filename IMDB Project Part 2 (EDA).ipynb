## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

## Load Datasets

```python
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2000.head()
```

```python
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df_2001.head()
```

## Concatenate DataFrames

```python
combined_df = pd.concat([df_2000, df_2001], ignore_index=True)
combined_df
```

### Make a Copy of the DataFrame

```python
eda_df = combined_df.copy()
```

## EDA

### Filtering the DataFrame

I want to know what movies generated revenue greater than zero or had any budget at all. To do this I will filter the combined_df for movies that were greater than zero in both features.

```python
revenue_filter = eda_df['revenue'] > 0
eda_df = eda_df.loc[revenue_filter,:]
eda_df.head()
```

```python
valid_rev = revenue_filter.value_counts()
valid_rev
```

The first question that I want to answer with visuals is how many movies generated revenue. I can do that with the filter I created above.

```python
plt.style.use('seaborn')
valid_rev.plot.bar(xlabel='Movies that Generated Reveune', ylabel='Count');
```

There were far less movies that generated revenue in the API data than there were movies that did.

### Filter out Movies with a zero Budget

For the rest of the dataset, I want to create visuals for movies that also had greater than zero budget. I will create another filter for that as well and update the DataFrame.

```python
budget_filter = eda_df['budget'] > 0
eda_df = eda_df.loc[revenue_filter & budget_filter,:]
eda_df.head()
```

### Number of Movies in Each Certification Category

Now I want to visualize the number of movies in each certification category.

```python
plt.style.use('seaborn')
sns.countplot(data=eda_df, x='certification', 
              order=['G', 'PG', 'PG-13', 'R', 'NR']);
```

From the countplot, it looks like R-rated movies were the highest number of movies while NR movies were the lowest.

### Average Revenue per Certification

An important piece of information to visualize is the amount of revenue generated by the different certification categories to understand what kind of movies generate the most amount of money.

```python
# using groupby to get the average revenue
avg_rev = eda_df.groupby(['certification'])['revenue'].mean().sort_values(ascending=False)
avg_rev
```

```python
fig, ax = plt.subplots(figsize=(10,8))
avg_rev.plot.bar()

price_fmt = mpl.ticker.StrMethodFormatter('${x:,.0f}')
ax.yaxis.set_major_formatter(price_fmt)
```

Even though G-rated movies were not as populace in the dataset, the average revenue was the greatest generated.

### Average Budget per Certification

On the flip side of revenue, I also want to visualize how the budget for each certification category differs. This can then be compared to the revenue to get a generalization of what the net profits are.

```python
# using groupby again to find the average
avg_bud = eda_df.groupby(['certification'])['budget'].mean().sort_values(ascending=False)
avg_bud
```

```python
fig, ax = plt.subplots(figsize=(10,8))
avg_bud.plot.bar()

price_fmt = mpl.ticker.StrMethodFormatter('${x:,.0f}')
ax.yaxis.set_major_formatter(price_fmt)
```

PG-rated movies appeared to have the highest budget on average, even though it only generated the second-highest revenue.
